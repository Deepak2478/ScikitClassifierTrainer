```python
import pandas as pd;
import numpy as np;
from sklearn.externals import joblib
from flask import Flask, jsonify
from flask import request
import json
from flask_cors import CORS, cross_origin
 
class IndicatorEnum:
    ERROR = 'Error'
    WARNIING = 'Warning'
    
    

class ValidationMessageObj:
    def __init__(self,IndicatorEnum,validationMsg):
        self.IndicatorEnum=IndicatorEnum;
        self.validationMsg=validationMsg;
                       
        
class StandardScaler:
          
    def __init__(self,dataset):
        global standardScaler;
        from sklearn.preprocessing import StandardScaler;
        standardScaler=StandardScaler();
        standardScaler=standardScaler.fit(X);
        
    def transform(self,dataset):
        return standardScaler.transform(dataset);
    
    def inverseTransform(self,dataset):
        return standardScaler.inverse_transform(dataset);
    

class KNNModel:
        
    def __init__(self,X_train,y_train,neighbours):
        global knn_Classifier;
        from sklearn.neighbors import KNeighborsClassifier;
        knn_Classifier=KNeighborsClassifier(n_neighbors = neighbours, metric = 'minkowski', p = 2)
        knn_Classifier=knn_Classifier.fit(X_train,y_train);
        
    def predict(self,X_test):
       return knn_Classifier.predict(X_test);
   
    def accuracy(self,y_true,y_pred):
        from sklearn.metrics import accuracy_score;
        return accuracy_score(y_true, y_pred);
    
    def getTrainedModel(self):
        return knn_Classifier;
    
class DecisionTreeClassifierModel:
    
    def __init__(self,X_train,y_train):
        global decisionClassifier;
        from sklearn.tree import DecisionTreeClassifier;
        decisionClassifier=DecisionTreeClassifier();
        decisionClassifier=decisionClassifier.fit(X_train,y_train);
        
    def predict(self,X_test):
       return decisionClassifier.predict(X_test);
   
    def accuracy(self,y_true,y_pred):
        from sklearn.metrics import accuracy_score;
        return accuracy_score(y_true, y_pred);
   
    def getTrainedModel(self):
        return decisionClassifier;
  
class RandomForestClassifierModel:
    
    def __init__(self,X_train,y_train,est):
        global classifier;
        from sklearn.ensemble import RandomForestClassifier
        classifier = RandomForestClassifier(n_estimators = est, criterion = 'entropy', random_state = 0)
        classifier.fit(X_train, y_train)

    def predict(self,X_test):
       return classifier.predict(X_test);
   
    def accuracy(self,y_true,y_pred):
        from sklearn.metrics import accuracy_score;
        return accuracy_score(y_true, y_pred);
    
    def getTrainedModel(self):
        return classifier;
   
    

class LabelEncoding:
    
    def __init__(self,dataset):
        global labelEncoder;
        from sklearn.preprocessing import LabelEncoder;
        labelEncoder=LabelEncoder();
        labelEncoder=labelEncoder.fit(dataset)
        
    def transform(self,y_dataset):
        return labelEncoder.transform(y_dataset);
    
    def inverseTransform(self,y_dataset):
        return labelEncoder.inverse_transform(y_dataset);
    

class SplittingDataset:
    
    def split_dataSet(self,X,y,test_Size,random_State):
        from sklearn.model_selection import train_test_split;
        return train_test_split(X,y,test_size=test_Size,random_state=random_State);


class ControllerHandler:
    def __init__(self,modelName,modelType,exportFlag,dataSet):
        self.modelName=modelName;
        self.modelType=modelType;
        self.exportFlag=exportFlag;
        self.dataSet=dataSet
        self.X_values=-1;
        self.y_values=-1;
        self.validationMsg=[];
        
        
    def validatingDataSet(self):
        try:
            #Basic validation on the dataset 
            arrayLen2D=np.shape(self.dataSet);
            
            lenRow=arrayLen2D[0];
            lenCol=arrayLen2D[1];
            
            if(lenRow<=1 or lenCol<=1):
                validationMessageObj=ValidationMessageObj(IndicatorEnum.ERROR,'Insufficient data in Rows or Columns');
                self.validationMsg.append(validationMessageObj);
                return self.X_values,self.y_value,self.validationMsg;
            
            self.X_values=self.dataSet[:,:-1];
            self.y_values=self.dataSet[:,-1:];
             
        except Exception as ex:
            validationMessageObj=ValidationMessageObj(IndicatorEnum.ERROR,ex);
            self.validationMsg.append(validationMessageObj);

        return self.X_values,self.y_values,self.validationMsg;
```

```python
app = Flask(__name__)
CORS(app)
@app.route('/trainingData', methods=['POST'])
def train():
    global labelEncoding
    jsonInputObj = request.json;
    modelName = jsonInputObj['modelName'];
    modelType= jsonInputObj['modelType'];
    exportFlag = jsonInputObj['exportFlag'];
    Array2DDataSet = jsonInputObj['DataSet'];
    dataSet=np.array(Array2DDataSet);

    controllerHandler=ControllerHandler(modelName,modelType,exportFlag,dataSet);
    X,y,validationRes=controllerHandler.validatingDataSet();
    y=y.ravel();

    if(len(validationRes)==0):
       
        labelEncoding=LabelEncoding(y);
        y=labelEncoding.transform(y);
        print(y);
        
        splittingDataset=SplittingDataset();
        X_train,X_test,y_train,y_test=splittingDataset.split_dataSet(X,y,0.25,1);
        
        accuracy=0; model_selection=None;
        
        if(modelType=='DTC'):
                decisionTreeClassifierModel=DecisionTreeClassifierModel(X_train,y_train);
                accuracy=decisionTreeClassifierModel.accuracy(y_test,decisionTreeClassifierModel.predict(X_test))
                model_selection=decisionTreeClassifierModel.getTrainedModel();
                
        if(modelType=='RFC'):
                randomForestClassifierModel=RandomForestClassifierModel(X_train,y_train,15);
                accuracy=randomForestClassifierModel.accuracy(y_test,randomForestClassifierModel.predict(X_test))
                model_selection=randomForestClassifierModel.getTrainedModel();
                
        if(modelType=='KNN'):
                 knnModel=KNNModel(X_train,y_train,3);
                 accuracy=knnModel.accuracy(y_test,knnModel.predict(X_test))
                 model_selection=knnModel.getTrainedModel();
        
        
        pickleDetails = '/home/payrollresults/Desktop/PayrollResults/exportedModels/'+modelName+'.pkl'
        joblib.dump(model_selection, pickleDetails) 
        json_data = json.dumps({'accuracy(%)' : (accuracy*100),
                                'location':pickleDetails});
        
    else:
       #print(validationRes[0].validationMsg);
       json_data = json.dumps({'Error Occured':validationRes[0].validationMsg})


    return json_data;
```
